```python
#Q2) 
def list_of_list(A, B):
    n = len(A)
    k = len(A[0])
    m = len(B[0])

    C = [[0 for _ in range(m)] for _ in range(n)]

    for i in range(n):
        for j in range(m):
            for p in range(k):
                C[i][j] += A[i][p] * B[p][j]

    return C
import numpy as np
def mul_numpy(A, B):
   return np.dot(A, B)
A = np.ones((50, 100))
B = np.ones((100, 200))
import time


start_time = time.time()
C = list_of_list(A, B)
end_time = time.time()
start_time1 = time.time()
C = mul_numpy(A, B)
end_time1 = time.time()
print(f"Time taken using list of lists: {end_time - start_time} seconds")
print(f"Time taken using numpy: {end_time1 - start_time1} seconds")
# from what i have understood numpy uses vectorization which runs in parallel rather than afterwards as in the case of list of lists. Hence numpy is generally faster.

```

```python
#Q3) using multi-pass method: Order=O(n)+O(n-1)+.....+O(n//2)=O(n^2)
import time
import numpy as np

def multipass(arr):
    n = len(arr)
    for _ in range(n // 2):
        # Find the maximum element
        max_val = max(arr)
        arr.remove(max_val)

    if n % 2 == 1:
        return max(arr)  # Median for odd length
    else:
        # For even length, return the average of the two middle elements
        first_middle = max(arr)
        arr.remove(first_middle)
        second_middle = max(arr)
        return (first_middle + second_middle) / 2

arr = np.random.randint(0, 100000, 10001).tolist()

start_time = time.time()
median_multipass = multipass(arr.copy())
end_time = time.time()
start_time1 = time.time()
median_numpy = np.median(arr)
end_time1 = time.time()

print(f"Median : {median_multipass}")
print(f"Time taken (multi-pass): {end_time - start_time} seconds")
print(f"Time taken (numpy): {end_time1 - start_time1} seconds")
# np.median generally follows O(n x log(n)), although quicksort follows just O(n).
```

```python
#Q8)
class Student:
    def __init__(self, roll_number, name, marks):
        self.roll_number = roll_number
        self.name = name
        self.marks = marks

    def __repr__(self):
        return f"Student(roll_number={self.roll_number}, name={self.name}, marks={self.marks})"

class Branch:
    def __init__(self, name):
        self.name = name
        self.students = []

    def add_student(self, student):
        self.students.append(student)

    def __repr__(self):
        return f"Branch(name={self.name}, students={self.students})"

class Year:
    def __init__(self, year):
        self.year = year
        self.branches = []

    def add_branch(self, branch):
        self.branches.append(branch)

    def __repr__(self):
        return f"Year(year={self.year}, branches={self.branches})"

class Database:
    def __init__(self):
        self.years = []

    def add_year(self, year):
        self.years.append(year)

    def __repr__(self):
        return f"Database(years={self.years})"

# Creating the database and adding data
database = Database()


year_2022 = Year(2022)
branch_1_2022 = Branch("Branch 1")
branch_1_2022.add_student(Student(1, "Student A", {"Maths": 100, "English": 70}))
branch_1_2022.add_student(Student(2, "Student B", {"Maths": 85, "English": 78}))

branch_2_2022 = Branch("Branch 2")
branch_2_2022.add_student(Student(1, "Student C", {"Maths": 90, "English": 88}))

year_2022.add_branch(branch_1_2022)
year_2022.add_branch(branch_2_2022)

database.add_year(year_2022)


year_2023 = Year(2023)
branch_1_2023 = Branch("Branch 1")
branch_1_2023.add_student(Student(1, "Student D", {"Maths": 95, "English": 80}))

branch_2_2023 = Branch("Branch 2")
branch_2_2023.add_student(Student(1, "Student E", {"Maths": 88, "English": 85}))

year_2023.add_branch(branch_1_2023)
year_2023.add_branch(branch_2_2023)

database.add_year(year_2023)


year_2024 = Year(2024)
branch_1_2024 = Branch("Branch 1")
branch_1_2024.add_student(Student(1, "Student F", {"Maths": 78, "English": 92}))

branch_2_2024 = Branch("Branch 2")
branch_2_2024.add_student(Student(1, "Student G", {"Maths": 84, "English": 76}))

year_2024.add_branch(branch_1_2024)
year_2024.add_branch(branch_2_2024)

database.add_year(year_2024)


year_2025 = Year(2025)
branch_1_2025 = Branch("Branch 1")
branch_1_2025.add_student(Student(1, "Student H", {"Maths": 91, "English": 89}))

branch_2_2025 = Branch("Branch 2")
branch_2_2025.add_student(Student(1, "Student I", {"Maths": 87, "English": 90}))

year_2025.add_branch(branch_1_2025)
year_2025.add_branch(branch_2_2025)

database.add_year(year_2025)


import pprint
pprint.pprint(database)
```

```python
#Q13)
import numpy as np
def my_argmin(arr):
    index = 0
    value = arr[0]
    for i in range(1, len(arr)):
        if arr[i] < value:
            value = arr[i]
            index = i
    
    return index
arr = np.array([3, 1, 4, -1, 5, 9, 2, 6, -5, 3, 5])
print("result:", my_argmin(arr))
print("NumPy argmin result:", np.argmin(arr))
```
